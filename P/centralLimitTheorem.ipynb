```python
#link for random generation https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.beta.html#numpy.random.Generator.beta
import pandas as pd
import numpy as np
```

```python
sample = np.random.default_rng().uniform(low=0,high = 1, size=100)
# sample = np.random.default_rng().exponential(scale=1, size=100)
pd.Series(sample).hist(bins=5, figsize=(5,3))
```

```python
nSamples = 1000
sampleSizes = [1, 5, 30, 1000]
allMeans = []
np.random.seed(1)
for n in sampleSizes:
    means = pd.Series(nSamples)
    for i in range(nSamples):
        sample = np.random.default_rng().exponential(scale=1, size=n)
        # sample = np.random.default_rng().uniform(low=0,high = 1, size=n)
        means[i] = pd.Series(sample).mean()
    allMeans.append(means)
    # myHist = means.hist(bins=5, figsize=(2,3))
    # myHist.set_xlabel('sampleSize='+str(n))
    
# for plotting all the means in one figure
import matplotlib.pyplot as plt
k = 0
fig, ax = plt.subplots(2, 2, figsize =(8, 7))
for i in range(0, 2):
    for j in range(0, 2):
        # Histogram for each x stored in allMeans
        ax[i, j].hist(allMeans[k], bins=10, density = True)
        ax[i, j].set_title(label = 'n='+str(sampleSizes[k]))
        k = k + 1
plt.show()
```
