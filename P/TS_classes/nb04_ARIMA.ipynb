## 1. Concepts Remembering

Again, the sequence of random variables $\{X_t: t = 1, 2, \dots \}$ is called a stochastic
process and serves as a model for an observed time series $\{x_t: t = 1, 2, \dots \}$. 

#### Random walk

Let $\{\varepsilon_t: t = 1, \dots\}$ be a sequence of independent, identically distributed random variables each with zero mean and constant variance. In the random walk stochastic process, one has $X_t = \sum_{i=1}^t\varepsilon_t$, i.e.:

$X_1 = \varepsilon_1$;

$X_2 = \varepsilon_1 + \varepsilon_2$;

$X_3 = \varepsilon_1 + \varepsilon_2 + \varepsilon_3$;

$X_4 = \varepsilon_1 + \varepsilon_2 + \varepsilon_3 + \varepsilon_4$.

So, one has $X_t = X_{t-1}+\varepsilon_t$. It leads to 
 
$\varepsilon_t = X_t - X_{t-1}$, assuming $X_0 = 0$

#### Stationarity

To make statistical inferences about the structure of a stochastic process on the basis of an observed record of that process, we must usually make some simplifying (and presumably reasonable) assumptions about that structure. The most important such
assumption is that of stationarity. The basic idea of stationarity is that the probability laws that govern the behavior of the process do not change over time. In a sense, the process is in statistical equilibrium. 

Specifically, a process $\{X_t\}$ is said to be strictly stationary if the joint distribution of $\{X_{t_1}, \dots, X_{t_n}\}$ is the same as the joint distribution of $\{X_{t_1-k}, \dots, X_{t_n-k}\}$
for all choices of time points $(t_1, \dots, t_n)$ and all choices of time
lag $k$. 

Thus, when $n = 1$ the (univariate) distribution of $X_t$ is the same as that of $X_{t − k}$ for
all $t$ and $k$; in other words, the $X$’s are (marginally) identically distributed. It then follows
that $E(X_t) = E(X_{t − k})$ for all $t$ and $k$ so that the mean function is constant for all time.
Additionally, $Var(X_t) = Var(X_{t − k})$ for all $t$ and $k$ so that the variance is also constant over
time.

Setting $n = 2$ in the stationarity definition we see that the bivariate distribution of $X_t$ and $X_s$ must be the same as that of $X_{t − k}$ and $X_{s − k}$ from which it follows that $Cov(X_t, X_s)= Cov(X_{t − k}, X_{s − k})$ for all $t, s$, and $k$.

#### General Linear Processes

Let $\{\varepsilon_t\}$ be a wtite noise, i.e. a sequence of identically distributed, zero-mean, independent random variables. $\{X_t\}$, is a general linear process, if it can be represented as a weighted linear combination of present and infinite past white noise terms as

$X_t = \varepsilon_{t} + \sum_{i=1}^\infty \theta_i \varepsilon_{t-i}$.

An important nontrivial example to which we will return often is the case where the
$\theta$’s form an exponentially decaying sequence: $\theta_i = \phi^i$. One has the initial condition that $\phi \in (−1, +1)$. Then, 

$X_t = \varepsilon_{t} + \sum_{i=1}^\infty\phi^i \varepsilon_{t-i}$.

#### Moving Average of order q - MA(q)

In the case where only a finite number, say $q$, of the $\theta$-weights are nonzero, we have what is
called a moving average process of order $q$:

$X_t = \varepsilon_{t} + \sum_{i=1}^q \theta_i \varepsilon_{t-i}$.

Moving average models were first considered by Slutsky (1927).

Ex. ($q=1$):

$X_0 = \varepsilon_0 = 0$

$X_1 = \varepsilon_1 + \theta_1 \varepsilon_{1-1}= 
\varepsilon_1 + \theta_1 \varepsilon_{0}=
\varepsilon_1 + \theta_1 X_0 = 
\varepsilon_1$

$X_2 = \varepsilon_2 + \theta_1 \varepsilon_{2-1}= 
\varepsilon_2 + \theta_1 \varepsilon_1=
\varepsilon_2 + \theta_1 X_1 = f(X_1) \longrightarrow 
\varepsilon_2 = X_2 - \theta_1 X_1$

$X_3 = \varepsilon_3 + \theta_1 \varepsilon_{3-1}= 
\varepsilon_3 + \theta_1 \varepsilon_2=
\varepsilon_3 + \theta_1 (X_2 - \theta_1 X_1) = f(X_2, X_1) \longrightarrow 
\varepsilon_3 = X_3 - \theta_1(X_2 - \theta_1 X_1)$

$X_4 = \dots = f(X_3, X_2, X_1) \longrightarrow X_t = f(X_{t-1}, X_{t-2}, \dots, X_1)$

Thus, a moving average process is in fact a autoregressive model that consideres all previous history of the process.  


### Autoregressive Processes  of order p - AR(p)

Autoregressive processes are as their name suggests—regressions on themselves. Specifically, a $p$th-order autoregressive process $\{X_t\}$ satisfies the equation

$X_t = \varepsilon_{t} + \sum_{i=1}^p \phi_i X_{t-i}$.

The current value of the series $X_t$ is a linear combination of the $p$ most recent past values
of itself plus an “innovation” term et that incorporates everything new in the series at
time $t$ that is not explained by the past values. Thus, for every $t$, we assume that $\varepsilon_t$ is
independent of $X_{t − 1}, X_{t − 2}, X_{t − 3}, \dots$ . Yule (1926) carried out the original work on
autoregressive processes.

### The Mixed Autoregressive Moving Average Model - ARMA(p,q)

If we assume that the series is partly autoregressive (of order $p$) and partly moving average (of order $q$), we
obtain a quite general time series model

$X_t = \varepsilon_{t} + \sum_{i=1}^p \phi_i X_{t-i} + \sum_{i=1}^q \theta_i \varepsilon_{t-i}$

### MODELS FOR NONSTATIONARY TIME SERIES

#### Stationarity Through Differencing

Perhaps the simplest nonstationary model is the random walk:

$X_t = X_{t-1}+\varepsilon_t$. One can rewrite 
$\nabla_t^1 = \varepsilon_t$, in which $\nabla_t^1 = X_t - X_{t-1}$ is thus the first difference of $X_t$. Thus, the random walk is easily extended to a more general model whose first difference is some stationary process—not just white noise. 

One can also have, in the same fashion, 

$\nabla_t^2 = 
\nabla_t^1 - \nabla_{t-1}^1 = 
(X_t - X_{t-1}) - (X_{t-1} - X_{t-2})$, the second difference of $X_t$. 

The $d$-th difference of $X_t$ (for $d=1, 2, \dots$) is thus given by $\nabla_t^d = 
\nabla_t^{d-1} - \nabla_{t-1}^{d-1}$, considering $\nabla_t^0 = X_t$.

#### ARIMA(p, d, q) Models

The time series $\{x_t\}$ is said to come from  an integrated autoregressive moving average (ARIMA) model if $\nabla^d_t$ is a stationary ARMA series. 
In the same line, if $\{\nabla^d_t\}$ follows an ARMA(p,q) model, we say that $\{x_t\}$ is an ARIMA($p,d,q$) model. Fortunately, for practical purposes, we can usually take $d = 1$ or at most 2 for achieving stationarity.

Ex.: ARIMA($p, 1, q$): 

$\nabla^1_t =  e_{t} + \sum_{i=1}^p \phi_i \nabla^1_{t-i} + \sum_{i=1}^q \theta_i e_{t-i}$, or, in terms of the observed series, target time series:

$x_t - x_{t-1} =  e_{t} + \sum_{i=1}^p \phi_i (x_{t-i}-x_{t-i-1}) + \sum_{i=1}^q \theta_i e_{t-i}$, which we may rewrite as:

$x_t  =  e_{t} + (\phi_1 + 1) x_{t-1} + (\phi_2 - \phi_1) x_{t-2} + \dots - \phi_p x_{t-p-1} + 
\sum_{i=1}^q \theta_i e_{t-i} \rightarrow $

$x_t  =  e_{t} + \sum_{i=1}^p (\phi_i - \phi_{i-1}) x_{t-i} - \phi_p x_{t-p-1} \sum_{i=1}^q \theta_i e_{t-i}
$, taking $\phi_0 = -1$. 

We call this  last equation the *difference equation form of the model*. It seems to be an ARMA ($p+1, q$).

#### Other Transformations

We have seen how differencing can be a useful transformation for achieving stationarity.
However, the logarithm transformation is also a useful method in certain circumstances.

If the standard deviation of the series is proportional to the level of the series,
then transforming to logarithms will produce a series with approximately constant variance over time. Also, if the level of the series is changing roughly exponentially, log-transformed series will exhibit a linear time trend. Thus, we might then want to take first differences. 

##### Power Transformations

A flexible family of transformations, the power transformations, was introduced by
Box and Cox (1964). For a given value of the parameter $\lambda$, the transformation is defined by 

$g_t = \frac{x_t^\lambda-1}{\lambda}$, if $\lambda \neq 0$. Otherwise, one has $g_t = \log x_t.$

The term $x_t^\lambda$ is the important part of the first expression, but subtracting 1 and dividing by $\lambda$ makes $g_t$ change smoothly as $\lambda$ approaches zero. In fact, a calculus argument shows that as $\lambda \rightarrow 0, (x_t^ \lambda − 1)/\lambda \rightarrow \log(x_t)$.

#### The Backshift (or Lag) Operator

Much of the time series literature use what is called the backshift
operator to express and manipulate ARIMA models. The backshift operator, denoted $B$,
operates on the time index of a series and shifts time back one time unit to form a new
series. 

$Bx_t = x_{t-1}$. 

The backshift operator is linear. Ex. for  MA(1) model: 

$x_t = e_t + \theta_1 e_{t-1} = 
e_t + \theta_1 B e_t = 
(1+B)e_t = 
\theta(B)e_t,$

in which  $\theta(B) = (1+B)$ is the named MA(1) characteristic polynomial function “evaluated” at point $B$.

Since $Bx_t$ is itself a time series, it is meaningful to consider $BBx_t = B^2x_t$. But clearly $BBx_t = Bx_{t − 1} = x_{t − 2}$. In general, one has $B^mx_t = x_{t-m}$.

#### The general ARIMA formula

**1. The general MA($q$) model:**

$x_t = e_t + \theta_1 e_{t-1} + \theta_2 e_{t-2} + \dots + \theta_q e_{t-q} = $

$ = e_t + \theta_1 B e_t + \theta_2 B^2e_t + \dots + \theta_q B^q e_t \rightarrow $

$ x_t = \theta(B) e_t$, in which $\theta(B)= \sum_{i=0}^q (\theta_i B^i)$ is the MA($q$) characteristic polynomial function evaluated at point $B$, $\theta_0 = 1$ and $B^0=1$.

**2. The general AR($p$) model:**

For autoregressive models AR($p$), we first move all of the terms involving $x$ to the
left-hand side

$x_t - \phi x_{t-1} + \phi_2 x_{t-2} + \dots + \phi_p x_{t-p}= e_t \rightarrow$

$ x_t - \phi_1 B x_t - \phi_2 B^2x_t - \dots - \phi_p B^p x_t = e_t  \rightarrow$

$ \sum_{i=0}^p (-\phi_i B^i) x_t = e_t,$ in which $\phi_0 = -1$ and $B^0=1$. 
One can also involve the AR characteristic polynomial function evaluated at $B$, $\phi (B) = \sum_{i=0}^p (-\phi_i B^i)$, leading to
$\phi(B)x_t = e_t$. 

Thus, one has

$ x_t = \theta(B) e_t$ from MA($q$) and

$\phi(B)x_t = e_t$ from AR($p$).

Remembering, the general ARMA($p,q$) model is expressed as

$x_t = e_{t} + \sum_{i=1}^p \phi_i x_{t-i} + \sum_{i=1}^q \theta_i e_{t-i}$. Rearranging, it can thus be concisely expressed as

$x_t - \sum_{i=1}^p \phi_i x_{t-i} = e_{t} +  \sum_{i=1}^q \theta_i e_{t-i} \rightarrow$

$\phi(B) x_t = \theta(B) e_t$

Differencing can also be conveniently expressed in terms of $B$: $\nabla x_t = x_t - B x_t = (1 – B) x_t$.

in the same way, the $d$-th difference is given by $\nabla^d x_t = (1 – B)^d x_t$. 

$\phi(B) (1-B)^d x_t = \theta(B) e_t$

### The Seasonal ARIMA - SARIMA

The extension to the seasonal model is natural. Let MA $(Q)[m]$ be the seasonal MA model of order $Q$ in which $m$ is the seasonal period.

**1. The general MA($Q$)[$m$] model:**
$%x_t = e_{t} + \sum_{i=1}^p \phi_i x_{t-i} + \sum_{i=1}^q \theta_i e_{t-i}$

$x_t = e_{t}  + \sum_{i=1}^Q \theta_i e_{t-m\times i} \rightarrow $

$x_t = e_{t}  + \sum_{i=1}^Q \theta_i B^{m\times i}e_t   \rightarrow$

$x_t = \theta(B)e_t$, in which $\theta(B) = \sum_{i=0}^Q \theta_i B^{m\times i}$ and $\theta_0=1$.

Thus, the general form of the MA($Q)[m]$ is the same o the MA($q$); though, in the MA($Q[m])$, the polynomial characteristic function involves $m$.

**2. The general AR($P$)[$m$] model:**


$x_t = e_{t}  + \sum_{i=1}^P \phi_i x_{t-m\times i} \rightarrow $

$x_t - \sum_{i=1}^P \phi_i B^{m\times i}x_t = e_{t}     \rightarrow$

$\phi(B)x_t = e_t$, in which $\phi(B) = -\sum_{i=0}^P \phi_i B^{m\times i}$ and $\phi_0=-1$. 

Thus, the general form of the AR($P)[m]$ is the same o the AR($p$); though, in the AR($P[m])$, the polynomial characteristic function involves $m$.

**3. The seasonal difference of period $m$:**

$\nabla^1_t[m] = x_t - x_{t-m}$ and, in general, 

$\nabla^D_t[m] = \nabla^{D-1}_t[m] - \nabla^{D-1}_{t-m}[m] $, in which $\nabla^{0}_t[m]  = x_t$.

**The general SARIMA model**

One has the SARIMA $(P, D, Q)(p, d, q)[m]$, 
in which  and $(P, D, Q)$ represents, in this order, the $m$-seasonal AR, I, and MA orders.

The traditional autocorrelation in a lag may be contamiated by intermediate lags. So a different function is needed. Such a function may be defined as the correlation between $X_t$
and $X_{t-k}$ after removing the effect of the intervening variables $X_{t-1}, \dots, X_{t-k + 1}$. This coefficient is called the *partial autocorrelation* at lag $k$ and will be denoted
by $\phi_{kk}$. If $X_t$ is a normally distributed variable, we can let:

$\phi_{kk} = Corr(X_t, X_{t-k}|X_{t-1}, \dots, X_{t-k+1})$

An alternative approach, not based on normality, can be developed in the following
way. Consider predicting $X_t$ based on a linear function of the intervening variables 

$\hat{X}_t = f(X_{t-1},\dots, X_{t-k + 1})$ = $\sum_{i=1}^{k-1}\beta_i X_{t-i}$, 

with the $\beta$’s chosen to
minimize the mean square error of prediction. If we assume that the $\beta$’s have been so
chosen and then think backward in time, **it follows from stationarity** that the best “predictor” of $X_{t − k}$ based on the same $X_{t − 1},\dots, X_{t − k + 1}$ will be 


$\hat{X}_{t-k} = \sum_{i=1}^{k-1}\beta_i X_{t-k+i}$. 

Ex. (Studyng the lag) $k=3$: 

$\hat{X}_t = f(X_{t-1}, X_{t-2})$ = $\beta_1 X_{t-1} + \beta_2 X_{t-2}$

$\hat{X}_{t-3} = f(X_{t-1}, X_{t-2})$ = 
$\beta_1 X_{t-3+1} + \beta_2 X_{t-3+2}$
= $\beta_1 X_{t-2} + \beta_2 X_{t-1}$.


Then, considering the error $\varepsilon_t = X_t - \hat{X}_t$, the partial autocorrelation function at lag $k$ is then
defined to be the correlation between the prediction errors; that is, $\rho_{kk} = Corr(\varepsilon_t, \varepsilon_{t-k})$ 

For normally distributed series, it can be shown that the two definitions coincide. By
convention, we take $\rho_{11} = 1$.

```python
#%pip install statsmodels
import statsmodels
import statsmodels.api as sm
#from statsmodels.tsa.stattools import coint, adfuller
fig = plt.figure(figsize=(9,6))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(nDj['e1RW(t)'][1:], lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(nDj['e1RW(t)'][1:], lags=40, ax=ax2)
```

```python
# acf, ci = sm.tsa.acf(endog, alpha=0.05)
def getSignificantLags(y, nLags = 5, alpha=0.05):
    pac, pac_ci = sm.tsa.pacf(x=y, nlags=nLags, alpha=alpha)
    pd.DataFrame(pac)[1:].plot(kind='bar', figsize=figsize, legend='pacf',
                               xlabel='lag', ylabel='pacf');
    significantLags = []
    for i in range(1, len(pac)):
        # print(pac[i], pac_ci[i][0], pac_ci[i][1])
        if pac[i] < pac_ci[i][0] - pac[i] or pac[i] > pac_ci[i][1] - pac[i]:
            significantLags.append(i)
    print('significantLags:', significantLags)
    return significantLags
e1_significantLags = getSignificantLags(y=nDj['e1RW(t)'][1:], nLags = 40, alpha=0.05)
```

```python
def appendLagVariables(yNm, significantLags, df, dropna=True, previousData = None):
    prefix = yNm.replace(')', '')
    df = df.copy()
    lowerIndex = 0
    x = df[yNm]
    if previousData is not None:
        x = pd.concat([previousData[yNm], x])
        lowerIndex = previousData.shape[0]
    for lag in significantLags:
        varNm = prefix+'-'+str(lag)+')'
        df[varNm] = x.shift(lag)[lowerIndex:]    
    if dropna:
        df.dropna(axis=0, inplace=True)
        
    display(df.head(2))
    return df
nDj.drop(['e1L(t-1)', 'e1RW(t-1)'], axis=1, inplace=True)
nDj = appendLagVariables(yNm='e1RW(t)', 
                         significantLags = e1_significantLags, 
                         df = nDj, 
                         dropna=True)
```

### Forecasting Performance

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_absolute_percentage_error
def TheilU(y_true, y_pred):
    fResiduals2 = y_true[1:].subtract(y_pred[1:])
    fResiduals2 = fResiduals2.multiply(fResiduals2)
    rwForecasts = y_true.shift(1)
    rwResiduals2 = y_true[1:].subtract(rwForecasts[1:])
    rwResiduals2 = rwResiduals2.multiply(rwResiduals2)
    theil = fResiduals2.sum()/rwResiduals2.sum()
    return theil
def computeForecastingPerformance(train, test, modelNms):
    """compute the forecasting performance of the predictors.

    Parameters:
    train (DataFrame): The train data frame with the target time series and respective 
    forecasts.
    test (DataFrame): The test data frame with the target time series and respective 
    forecasts.
    modelNms (string): The names of the predictor columns. 

    Returns:
    peformance (DataFrame): The perfomance in train and test sets, in terms of 
    MAE, MAPE, R2, and TheilU.
    """
    MAE_train = []; MAPE_train = []; R2_train = []; Theil_train = []
    MAE_test = []; MAPE_test = []; R2_test = []; Theil_test = []
    train.dropna(inplace=True)
    target_train = train['y']
    target_test = test['y']
   
    for modelNm in modelNms:
        forecasts_train = train[modelNm]
        forecasts_test = test[modelNm]
        
        MAE_train.append(mean_absolute_error(y_true = target_train, y_pred= forecasts_train))
        MAPE_train.append(mean_absolute_percentage_error(y_true = target_train, y_pred= forecasts_train))
        R2_train.append(r2_score(y_true = target_train, y_pred= forecasts_train))
        Theil_train.append(TheilU(y_true = target_train, y_pred= forecasts_train))

        MAE_test.append(mean_absolute_error(y_true = target_test, y_pred= forecasts_test))
        MAPE_test.append(mean_absolute_percentage_error(y_true = target_test, y_pred= forecasts_test))
        R2_test.append(r2_score(y_true = target_test, y_pred= forecasts_test))
        Theil_test.append(TheilU(y_true = target_test, y_pred= forecasts_test))

    rDf = pd.DataFrame({'model': modelNms, 
                       'MAE_tr': MAE_train, 
                       'MAE_te': MAE_test, 
                       'MAPE_tr': MAPE_train,
                       'MAPE_te':MAPE_test,
                       'R2_tr': R2_train,
                       'R2_te':R2_test,
                       'Theil_tr': Theil_train,
                       'Theil_te':Theil_test
                       })
    print(" *********** Rank via train set-based Accuracy and RMSE-based Cross-validation *********** ") 
    display(rDf.sort_values(by="Theil_te", ascending = True))
        
computeForecastingPerformance(train=nDj, test = test, 
                              modelNms = ['p1L', 'p1RW', 'p1RW+p2', 'p1RW+p2+p3XGB'])    
```
