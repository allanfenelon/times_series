```python
#VARIÁVEIS ALEATÓRIAS DISCRETAS
#link for random generation https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.beta.html#numpy.random.Generator.beta
#link for probability distribution functions https://docs.scipy.org/doc/scipy-0.16.1/reference/stats.html
import pandas as pd
# !pip install scipy
```

```python
#Uniforme Discreta
x = list(range(1, 7))#exemplo do dado, Omega_x = (1, 2, ...., 6)
# print(x)
Omega_x_card = len(x)
Px = [1/Omega_x_card]*Omega_x_card#probability mass function
# print(Px)
Rx = []#Rx(x_i) = P(X>x_i) = 1 - P(X<=x_i) 
for x_i in x:
    Rx.append(1 - x_i/Omega_x_card)#reliability: P(X>x_i) = 1 - P(X<=x_i) 
# print(Rx)
def printDistributions(x, Px, Rx, distName):
    dic = {'x':x, 'P(x)':Px, 'R(x)': Rx}
    # print(dic)
    df = pd.DataFrame(data=dic)
    print("************* " + distName + " PROBABILITY DISTRIBUTIONS *************")
    display(df)
    df.plot.bar(x = 'x', y = 'P(x)', figsize = (3,3))
    df.plot.bar(x = 'x', y = 'R(x)', figsize = (3,3))
printDistributions(x, Px, Rx, "Discrete Uniform")
```

```python
#Geométrica
from scipy.stats import geom
# import matplotlib.pyplot as plt
#fig, ax = plt.subplots(2, 1)
p = .1#probabilidade de sucesso em uma tentativa
x = list(range(1,21))
dist = geom(p)#Omega_X = {1, 2, 3, ...}
Px = dist.pmf(x)#probability mass function: P(X=x)
Rx = dist.sf(x)#Reliability/Survival function: P(X>x) = 1 - P(X<=x)
printDistributions(x, Px, Rx, "Geometric")
```

```python
#Binomial
from scipy.stats import binom
p = .9#probabilidade de sucesso em uma tentativa
n = 20#tamanho amostral
x = list(range(0,(n+1)))#Omega_x = {0, 1, 2, ...., n}
dist = binom(n, p)#Omega_X = {1, 2, 3, ..., n}
Px = dist.pmf(x)#probability mass function: P(X=x)
Rx = dist.sf(x)#Reliability/Survival function: P(X>x) = 1 - P(X<=x)
printDistributions(x, Px, Rx, "Binomial")
# choose(1e+100,30)
```

```python
#Hipergeométrica
from scipy.stats import hypergeom
N=20#tamanho da população (M in Python)
K=10#nº de sucessos na população (n in Python)
print("Probability success in one attempt = " + str(K/N))
n=6#tamanho da amostra (N in Python)
x = list(range(0,(min(n, K)+1)))#Omega_X
dist = hypergeom(M = N, n = K, N = n)#Omega_X = {0, 1, ..., min(n, K)}
Px = dist.pmf(x)#probability mass function: P(X=x)
Rx = dist.sf(x)#Reliability/Survival function: P(X>x) = 1 - P(X<=x)
printDistributions(x, Px, Rx, "Hypergeometric")
```

```python
#Poisson
from scipy.stats import poisson
myLambda= 10#taxa (nº médio) de ocorrências por unidade de medida (mu in Python)
x = list(range(0,20))#
dist = poisson(mu = myLambda)#Omega_X = {0, 1, ...}
Px = dist.pmf(x)#probability mass function: P(X=x)
Rx = dist.sf(x)#Reliability/Survival function: P(X>x) = 1 - P(X<=x)
printDistributions(x, Px, Rx, "Poisson")
```

```python
#Mistura entre Poisson e Binomial ...
from scipy.stats import poisson
from scipy.stats import binom
#Y: nº de novos casos de Covid-19 por dia em dado local
#Y~Poisson(lambda=1.2)
Y_dist = poisson(mu = 1.2)#Omega_Y = {0, 1, ...}
pY_greaterOrEqualTo3 = Y_dist.sf(2)#P(Y>2) = P(Y>=3) 
print("P(Y>=3) = ", pY_greaterOrEqualTo3)
n=10
#X: nº de locais em que ocorre ao menos 3 casos de Covid-19 por dia, de uma
#amostra de (n=)10 locais
#X~binomial(n=10, p=P(Y>=3))
x = list(range(0, n+1))
X_dist = binom(n, pY_greaterOrEqualTo3)
Px = X_dist.pmf(x)#probability mass function: P(X=x)
Rx = X_dist.sf(x)#Reliability/Survival function: P(X>x) = 1 - P(X<=x)
printDistributions(x, Px, Rx, "Poisson-based Binomial")
```
