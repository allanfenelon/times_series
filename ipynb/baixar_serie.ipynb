$\textbf{1. Tesla, Inc. (TSLA)} \\$
A Tesla é uma das líderes globais na produção de veículos elétricos e soluções de armazenamento de energia. Fundada por Elon Musk, a empresa é conhecida por seus carros elétricos de alto desempenho, como o Model S, Model 3, Model X e Model Y. Além de veículos elétricos, a Tesla também produz baterias e sistemas de energia solar. \\

$\textbf{2. NIO Inc. (NIO)} \\$
A NIO é uma fabricante de veículos elétricos premium com sede na China. A empresa é conhecida por seus SUVs elétricos e sedans, como o ES8, ES6 e o EC6. A NIO também se destaca por seu modelo de substituição de bateria, oferecendo um serviço onde os clientes podem trocar as baterias de seus veículos em estações de troca rápida. \\

$\textbf{3. Li Auto Inc. (LI)} \\$
Li Auto é uma fabricante chinesa de veículos elétricos e híbridos plug-in. A empresa é conhecida por seus SUVs híbridos, como o Li ONE. Li Auto se concentra em fornecer veículos elétricos com tecnologias avançadas e uma extensa rede de recarga e suporte. \\

$\textbf{4. XPeng Inc. (XPEV)} \\$
XPeng é outra fabricante chinesa de veículos elétricos, focada em carros inteligentes e conectados. A empresa oferece modelos como o G3 (SUV) e o P7 (sedan). XPeng é conhecida por suas inovações tecnológicas e sistemas avançados de assistência ao motorista. \\

$\textbf{5. BYD Company Limited (BYDDY)} \\$
A BYD é uma das maiores fabricantes de veículos elétricos e baterias do mundo. Com sede na China, a empresa produz uma ampla gama de veículos elétricos, incluindo ônibus e carros de passageiros, além de baterias e tecnologias de energia renovável. A BYD tem um forte foco em soluções sustentáveis e é uma das líderes no mercado de veículos elétricos na China. \\

$\textbf{6. Rivian Automotive Inc. (RIVN)} \\$
Rivian é uma empresa americana focada na produção de veículos elétricos aventureiros e robustos, como a picape R1T e o SUV R1S. Fundada em 2009, a Rivian tem atraído bastante atenção com seu design inovador e foco em um mercado de nicho para veículos elétricos off-road e utilitários. \\

$\textbf{7. Lucid Motors Inc. (LCID)} \\$
Lucid Motors é uma fabricante de veículos elétricos de luxo com sede na Califórnia. A empresa é conhecida por seu sedã elétrico de alto desempenho, o Lucid Air, que compete com modelos premium como o Tesla Model S. Lucid se destaca por seu foco em tecnologia avançada e design sofisticado.


```python
import yfinance as yf
import pandas as pd

# Definir o período de 1 ano
start_date = "2020-10-01"
end_date = "2024-10-01"

# Tickers das empresas
tickers = ['TSLA', 'NIO', 'LI', 'XPEV', 'BYDDY', 'RIVN', 'LCID']

# Função para baixar os dados mensais de uma ação
def download_stock_data(ticker, start, end):
    stock_data = yf.download(ticker, start=start, end=end, interval='1d')
    stock_data['Ticker'] = ticker + '_diario'
    return stock_data

# Caminho base para salvar os arquivos
base_path = '../bds/'

# Baixar dados e salvar arquivos CSV individuais
for ticker in tickers:
    data = download_stock_data(ticker, start_date, end_date)
    save_path = f'{base_path}{ticker}.csv'
    data.to_csv(save_path, index=True)
    print(f"Dados de {ticker} salvos em: {save_path}")
```

```python
def get_listing_exchange(ticker):
    stock = yf.Ticker(ticker)
    info = stock.info
    return info.get('exchange', 'Não disponível')

# Verificar a bolsa de listagem de cada ticker
for ticker in tickers:
    exchange = get_listing_exchange(ticker)
    print(f"{ticker} está listado na: {exchange}")
```

```python
import matplotlib.pyplot as plt

for ticker in tickers:
    data = download_stock_data(ticker, start_date, end_date)
    
    # Plotar os dados de fechamento
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], marker='o', label=ticker)
    plt.title(f'Preço de Fechamento Mensal de {ticker}')
    plt.xlabel('Data')
    plt.ylabel('Preço de Fechamento (USD)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    
    # Mostrar o gráfico
    plt.show()
```
